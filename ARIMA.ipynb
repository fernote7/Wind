```python
from __future__ import print_function
__author__ = 'Fernando'

%pylab inline

import os
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

#os.chdir("C:\Users\Fernando\Dropbox (Pessoal)\Fernando\Dissertacao\Scripts")




wind = pd.read_csv('potenMERRA.csv', sep=';', encoding='latin1', )
print(wind)

```

```python
wind.plot(figsize=(14,10))
plt.axis([1, 760, 0, 2700])
plt.xlabel('Semana');
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(wind.values.squeeze(), lags=90, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(wind, lags=90, ax=ax2)
```

```python
wind.index = pd.date_range(start= '2000-01-07', periods=747, freq="7D")

#print(wind.index)

arima_mod302 = sm.tsa.ARIMA(wind, (3,0,2), dates=wind.index).fit()
print (arima_mod302.params)

arma_mod32 = sm.tsa.ARMA(wind, (3,2), dates=wind.index).fit()
print (arma_mod32.params)
```

```python
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)
ax = arma_mod32.resid.plot(ax=ax);
plt.axis()
plt.xlabel('Residuos');
```

```python
predict_wind = arma_mod32.predict(650, 747, dynamic=True)
a = predict_wind.to_csv("predict_wind.csv", sep=";", header=False, index_label=['date','prediction'])

print (predict_wind)

```

```python
ax = wind.plot(figsize=(12,8))

ax = predict_wind.plot(ax=ax, style='r-', label='Dynamic Prediction');
ax.legend();

ax.axis();
plt.xlabel('Potencia')
plt.ylabel('Semanas');
```
